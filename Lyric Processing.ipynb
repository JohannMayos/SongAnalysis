```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


file_path = 'C:/Users/jmayo/SongAnalysis/h_artists_songs.csv'
df = pd.read_csv(file_path)
```

```python
df.head()

df.info()

df.isnull().sum()
```

```python
df_unique_lyrics = df.drop_duplicates(subset='Lyrics')

print(f"Total de letras únicas: {df_unique_lyrics.shape[0]}")

df_unique_urls = df.drop_duplicates(subset='Lyrics_URL')

print(f"Total de URLs únicas: {df_unique_urls.shape[0]}")
```

```python

recordings_per_year = df['Year'].value_counts().sort_index()

plt.figure(figsize=(12, 6))
recordings_per_year.plot(kind='bar')
plt.title('Número de Gravações por Ano')
plt.xlabel('Ano')
plt.ylabel('Número de Gravações')
plt.show()
```

```python

recordings_per_artist = df['Artist_Name'].value_counts()

top_10_artists = recordings_per_artist.head(10)
plt.figure(figsize=(12, 6))
top_10_artists.plot(kind='bar')
plt.title('Top 10 Artistas por Número de Gravações')
plt.xlabel('Artista')
plt.ylabel('Número de Gravações')
plt.xticks(rotation=45)
plt.show()
```

```python

df['Lyrics_Length'] = df['Lyrics'].str.len()
print(df['Lyrics_Length'].describe())


plt.figure(figsize=(12, 6))
df['Lyrics_Length'].hist(bins=50)
plt.title('Distribuição do Tamanho das Letras')
plt.xlabel('Tamanho da Letra')
plt.ylabel('Frequência')
plt.show()
```
